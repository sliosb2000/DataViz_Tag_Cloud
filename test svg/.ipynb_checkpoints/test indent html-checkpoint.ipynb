# creating the SVG, creating html and embeding the svg

### importing the libraries and data

```python
from email import message
import os

from os import path
from typing import TextIO

from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
# get data directory (using getcwd() is needed to support running example in generated IPython notebook)
d = path.dirname(__file__) if "__file__" in locals() else os.getcwd()

# Read the whole text.
text = open(path.join(d, 'input.txt'),encoding="utf-8").read()
print(text)
```

### generate the wordcloud

```python
### # generates the word cloud with arguments
wordcloud = WordCloud(background_color = 'white', max_font_size=40).generate(text)
```

### plotting the wordcloud

```python
plt.figure(figsize = (8, 8), facecolor = None)
plt.axis("off")
plt.imshow(wordcloud)
```

### Saving the SVG

```python
wordcloud_svg = wordcloud.to_svg(embed_font=True)
f = open("cloud.svg","w+")
f.write(wordcloud_svg)
info = wordcloud_svg
f.close()

with open('cloud.svg', 'r') as file:
    pass
```

### writing the HTML

```python
# We need to make an HTML Generator in python
# the idea is to use a HTML page to format the interaction with the SVG
# so by importing the SVG into the HTML document and by making the CSS for all <text> items
# All the words will react in the same way (depending on what we choose them to do)
f = open('index.html','w')

Top = """
<html>
    <head>
    <style>
    text:focus,text:hover {
        font-weight: bold;
    }
    </style>
    </head>
    <body>
"""
Bottom ="""
    </body>
</html>
"""
#f.write(Top + info + Bottom) # not usefull in this case in a notebook
html = Top + info + Bottom
#f.close 
```
